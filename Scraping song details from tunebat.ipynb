```python
def get_songs(x):
    url = "https://tunebat.com/Search?q={}".format(x)
    r=requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    
    artist=[]
    key=[]
    camelot=[]
    bpm=[]
    track_name=[]
    
    for song in soup.find_all('div', attrs={'class':'row main-row'}):
        try:
            artist.append(song.find('div', attrs={'class': 'row search-artist-name'}).text)
        except:
            artist.append(np.nan)
        try:
            track_name.append(song.find('div', attrs={'class': "row search-track-name"}).text)
        except:
            track_name.append(np.nan)
        try:
            key.append(song.find_all('div', attrs={'class':"col-md-4 col-sm-4 col-xs-4"})[0].text)
        except:
            key.append(np.nan)
        try:
            camelot.append(song.find_all('div', attrs={'class':"col-md-4 col-sm-4 col-xs-4"})[1].text)
        except:
            camelot.append(np.nan)
        try:
            bpm.append(song.find_all('div', attrs={'class':"col-md-4 col-sm-4 col-xs-4"})[2].text.strip('Camelot'))
        except:
            bpm.append(np.nan)
        
    scraped={'artist':artist,
        'track_name':track_name,
        'key':key,
        'camelot':camelot,
        'bpm':bpm}
    
    df=pd.DataFrame(scraped).dropna()
        
    return df
```
